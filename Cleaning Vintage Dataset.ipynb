```python
import pandas as pd
url='https://bit.ly/vintage-auctions'
df=pd.read_csv(url)
df
```

```python
df.dtypes
```

```python
df['daytime']=pd.to_datetime(df['time'])
```

```python
colonne_temporaire=df['daytime'].dt.hour*100+df['daytime'].dt.minute
colonne_temporaire
```

```python
bins= [0,324,1612,2211,2400]
labels=['NUIT1','MATIN','AP.MIDI','NUIT2']
```

```python
daytimetable=pd.cut(colonne_temporaire,bins=bins,labels=labels)
daytimetable
```

```python
df['daytime']=daytimetable
df
```
